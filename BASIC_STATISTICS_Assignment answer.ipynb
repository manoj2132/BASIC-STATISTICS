```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.read_csv('/content/sales_data_with_discounts.csv')
df
```

●	**Histograms:**

```python

# Example: Distribution of Average Price
df['Avg Price'].hist(bins=40)
plt.xlabel('Avg Price')
plt.ylabel('Frequency')
plt.title('Distribution of Avg Price')
plt.show()
```

```python


# Assuming df is your DataFrame with numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Plotting histograms for each numerical column
df[numerical_columns].hist(bins=30, figsize=(15, 10), layout=(len(numerical_columns) // 3 + 1, 4))
plt.tight_layout()
plt.show()
```

```python
 #volumes:The Volume column might show a right-skewed distribution if most sales volumes are low, with a few instances of very high sales volumes.
 #Avg Price:A right-skewed distribution would indicate that most products are low-priced with some high-priced outliers.
 #Total Sales Value:If there are two peaks, it could suggest different categories of sales (e.g., high-value and low-value products).
 #Discount Rate (%):If the histogram is left-skewed, it indicates that most items have a high discount rate with a few items at lower rates.
```

●	**Bar Chart Analysis for Categorical Column:**

```python
categorical_columns=['Day','SKU','City','BU','Brand','Model']
plt.figure(figsize=(15, 10))

for i, column in enumerate(categorical_columns, 1):
  #The enumerate function provides both the index (i) and the column name (column).
  #The 1 indicates that the indexing starts at 1 instead of the default 0.
    plt.subplot(len(categorical_columns) // 3 + 1, 3, i)  # Create subplots in a grid
    sns.countplot(data=df, x=column, palette='viridis')  # Create bar plot
    plt.title(f'Bar Plot of {column}')
    plt.xticks(rotation=45)
    print(df[column].value_counts(normalize=True) * 100)


plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply the scaler to the numerical columns
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Display the standardized DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Plot histograms of the standardized numerical columns
df[numerical_columns].hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Example: Assume 'df' is your DataFrame and it contains categorical columns like 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model'
categorical_columns = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

# Apply one-hot encoding to categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Display a portion of the transformed dataset
print(df_encoded.head())
```
